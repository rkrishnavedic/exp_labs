```python
import math
import numpy as np
import matplotlib.pyplot as plot

#Lagged fibonacci Generator

#given values for lcg
m=2048
b=1
a=1229
seed=0

x=seed

u_prev=[]
u_next=[]

#part a of problem 1

for _ in range(17):#generating first 17 u_i values
    
    u=x/m
    
    x=(a*x+b)%m
    
    if(x==seed):
        break
    
    u=float("{:.2f}".format(u))
    u_prev.append(u)
    u=x/m;
    u=float("{:.2f}".format(u))
    u_next.append(u)
    
#part b of problem 1 
#max_length

max_length=[1e3, 1e4, 1e5]

u_previous=u

print("---------\n Now upto range : 1000")

for _ in range(1000):
    
    size=len(u_prev)
    
    u=u_prev[size-17]-u_prev[size-5]
    u_prev.append(u_next[size-1])
    
    if(u<0):
        u=u+1.0
    
    u_next.append(u)
    
    

plot.figure(figsize=(10,10))
plot.scatter(u_prev,u_next,color='g', ec='y')
plot.show()


plot.hist(u_prev, bins=20, color='r', ec='b')
       
plot.xlabel('intervals of size 0.05')
plot.ylabel('frequency')
plot.title("For 1000 values of U_i")
plot.show()

        
print("--------\n Now upto range: 10000")

for _ in range(1000,10000,1):
    
    size=len(u_prev)
    
    u=u_prev[size-17]-u_prev[size-5]
    u_prev.append(u_next[size-1])
    
    if(u<0):
        u=u+1.0
    
    u_next.append(u)
    
    

plot.figure(figsize=(10,10))
plot.scatter(u_prev,u_next,color='g', ec='y')
plot.show()

plot.hist(u_prev, bins=20, color='r', ec='b')
plot.xlabel('intervals of size 0.05')
plot.ylabel('frequency')
plot.title("For 10000 values of U_i")
plot.show()


print("---------\n Now upto range: 100000")

for _ in range(10000,100000,1):
    
    size=len(u_prev)
    
    u=u_prev[size-17]-u_prev[size-5]
    u_prev.append(u_next[size-1])
    
    if(u<0):
        u=u+1.0
    
    u_next.append(u)
    
    

plot.figure(figsize=(10,10))
plot.scatter(u_prev,u_next,color='g', ec='y')
plot.show()

plot.hist(u_prev, bins=20, color='r', ec='b')
plot.xlabel('intervals of size 0.05')
plot.ylabel('frequency')
plot.title("For 100000 values of U_i")
plot.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random as rd

x=[];

N=200000;

for _ in range(10000):
    if(_%2==1):
        x.append(_);

total=5000;
inc=1/total;

q=[0]*5001;
f=[0]*5000;

for _ in range(1,5001):
    q[_]=q[_-1]+inc;

freq_data=[]

some_data_upto=10;
    
for _ in range(N):
    
    r=(rd.uniform(0,1))*total;
    freq_data.append(r);
    fl=math.ceil(r);
    if(fl!=0):
        f[fl-1]+=1;
        
    if some_data_upto>=0:
        some_data_upto-=1;
        print(str(fl*2+1), end=" ");
    
    #print("how?")

plt.hist(freq_data,bins= 50, color='#ff8833', ec='white');
plt.xlabel("x-represent k where q_(k-1) < U <= q_k")
plt.ylabel("y represent the freqency of numbers in that bin")
plt.show();


```

```python
import math
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt

m=2048
b=1
a=1229
seed=0

x=seed

u_prev=[]
u_next=[]

for _ in range(17):#generating first 17 u_i values
    
    u=x/m
    
    x=(a*x+b)%m
    
    if(x==seed):
        break
    
    u=float("{:.2f}".format(u))
    u_prev.append(u)
    u=x/m;
    u=float("{:.2f}".format(u))
    u_next.append(u)
    
#part b of problem 1 
#max_length

max_length=[1e3, 1e4, 1e5]

u_previous=u

#here we have computed u_i values for interval [0,1] from problem 1

#max_range
N=100000

for _ in range(N):
    
    size=len(u_prev)
    
    u=u_prev[size-17]-u_prev[size-5]
    u_prev.append(u_next[size-1])
    
    if(u<0):
        u=u+1.0
    
    u_next.append(u)
    
#Problem 2

theta=0.5
parts=100

X=[]

print("For N=1000 U in /U[0,1]")


for _ in range(1000):
    value = 0.5-0.5*math.cos((math.pi)*u_next[_])
    X.append(value)

 
mn=min(X)
mx=max(X)


def f1(t):
    if t==0:
        return 100
    if t==1:
        return 100
    return (1/np.pi)*(1/np.sqrt(t*(1-t)))*(10)

f=np.vectorize(f1)


for _ in range(1):
    
    #print("For N=1000 Samples")
    
    mean_X=stat.mean(X)
    variance_X=stat.variance(X)
    print(mean_X, "--",variance_X)
    
    t1=np.arange(mn, mx, (mx-mn)/100)
    plt.figure(figsize=(15,15))
    
#     plt.hist(X, bins=100, color='r', ec='b')
#     plt.xlabel('intervals of size 0.02')
#     plt.ylabel('frequency')
#     plt.title("For 1000 values of X")
    #plt.show()
    
    y,edges = np.histogram(X, t1)
    centers = 0.5*(edges[1:]+ edges[:-1])
    plt.plot(centers,y,'b-*')
    #plt.show()
    plt.title("mean = "+str(mean_X) + ", variance = "+ str(variance_X))
    
    plt.plot(t1, f(t1), 'r')
    plt.show()
    

print("For N=10000 U in /U[0,1]")


for _ in range(1000,10000,1):
    value = 0.5-0.5*math.cos((math.pi)*u_next[_])
    X.append(value)

 
mn=min(X)
mx=max(X)

def f1(t):
    if t==0:
        return 1000
    if t==1:
        return 1000
    return (1/np.pi)*(1/np.sqrt(t*(1-t)))*(100)

f=np.vectorize(f1)

for _ in range(1):
    
    #print("For N=1000 Samples")
    
    mean_X=stat.mean(X)
    variance_X=stat.variance(X)
    print(mean_X, "--",variance_X)
    
    t1=np.arange(mn, mx, (mx-mn)/100)
    plt.figure(figsize=(15,15))
    
#     plt.hist(X, bins=100, color='r', ec='b')
#     plt.xlabel('intervals of size 0.02')
#     plt.ylabel('frequency')
#     plt.title("For 1000 values of X")
    #plt.show()
    
    y,edges = np.histogram(X, t1)
    centers = 0.5*(edges[1:]+ edges[:-1])
    plt.plot(centers,y,'b-*')
    #plt.show()
    plt.title("mean = "+str(mean_X) + ", variance = "+ str(variance_X))
    
    plt.plot(t1, f(t1), 'r')
    plt.show()
    

print("For N=100000 U in /U[0,1]")


for _ in range(10000,100000,1):
    value = 0.5-0.5*math.cos((math.pi)*u_next[_])
    X.append(value)

 
mn=min(X)
mx=max(X)

def f1(t):
    if t==0:
        return 10000
    if t==1:
        return 10000
    return (1/np.pi)*(1/np.sqrt(t*(1-t)))*(1000)

f=np.vectorize(f1)

for _ in range(1):
    
    #print("For N=1000 Samples")
    
    mean_X=stat.mean(X)
    variance_X=stat.variance(X)
    
    t1=np.arange(mn, mx, (mx-mn)/100)
    plt.figure(figsize=(15,15))
    
#     plt.hist(X, bins=100, color='r', ec='b')
#     plt.xlabel('intervals of size 0.02')
#     plt.ylabel('frequency')
#     plt.title("For 1000 values of X")
    #plt.show()
    
    y,edges = np.histogram(X, t1)
    centers = 0.5*(edges[1:]+ edges[:-1])
    plt.plot(centers,y,'b-*')
    plt.title("mean = "+str(mean_X) + ", variance = "+ str(variance_X))
    #plt.show()
    
    plt.plot(t1, f(t1), 'r')
    plt.show()
    

















```
